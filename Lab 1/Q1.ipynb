## Question 1: Basic Representation of Signals

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal
```

```python
x = np.linspace(-12.5, 12.5, 101)
plt.figure()
```

```python
# These are the signals functions
y1 = np.sin(x)
y2 = np.cos(x)
y3 = signal.unit_impulse(101)
y4 = x > 0
y5 = (x > -3) * (x < 3)  # Change width if want
y6 = signal.sawtooth(x, 0.5)
y7 = np.exp(x)
y8 = signal.sawtooth(x)
y9 = 2 * (x > 0) - 1
y10 = np.sinc(x)  # sin(x) / x
```

```python
wave_name = [
    'Sine wave', 'Cosine wave', 'Unit Impulse wave', 'Unit step wave',
    'Square wave', 'Triangular wave', 'Exponential wave', 'Sawtooth wave',
    'Signum function', 'Sinc fucntion'
]
y = [y1, y2, y3, y4, y5, y6, y7, y8, y9, y10]

plt.figure(figsize=(10, 10))
for i, yi in enumerate(y):
    plt.subplot(5, 2, i + 1)
    plt.plot(x, yi)
    plt.title(wave_name[i])
    plt.ylabel('Amplitude')
    plt.xlabel('Time')
plt.tight_layout()
plt.show()
```
