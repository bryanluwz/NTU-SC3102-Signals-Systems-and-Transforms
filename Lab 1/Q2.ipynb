### Question 2.1: Sampling Theorem

```python
# Frequency of sine wave @ 100 Hz
wave_freq = 100

t1 = np.linspace(0, 0.1, 200)
y1 = np.sin(2 * np.pi * wave_freq * t1 +
            np.pi / 6)  # y(t) = A sin(2 pi * 100t + pi / 6)

sampling_freqs = [wave_freq * 2 - 30, wave_freq * 2, wave_freq * 2 + 69]

# Plot the since wave sampling at different frequencies and plot original in a new figure
# Plot it tightly together
for i in range(len(sampling_freqs)):
	plt.figure(i + 1)
	t2 = np.linspace(0, 0.1, int(1 / sampling_freqs[i] * 10000))
	y2 = np.sin(2 * np.pi * wave_freq * t2 +
				np.pi / 6)  # y(t) = A sin(2 pi * 100t + pi / 6)
	plt.plot(t1, y1, label='Original')
	plt.plot(t2, y2, label='Sampled')
	plt.xlabel('Time (s)')
	plt.ylabel('Amplitude')
	plt.title('Sampling Frequency = ' + str(sampling_freqs[i]) + ' Hz')
	plt.legend()
	plt.tight_layout()
	
```

### Question 2.2: Addition of two discrete time signals

```python
def gcd(x, y):
	"""
	Finds the greatest common divisor of two numbers
	"""
	while(y):
		x, y = y, x % y
	return x

def lcm(x, y):
	"""
	Finds the least common multiple of two numbers
	"""
	return (x * y) // gcd(x, y)
```

```python
Phi = 0
Fs = 60
sTime = 0
eTime = 1

# Y1 = A1 * cos(2 * pi * F1 * t + Phi)
A = 0.5
F1 = 10

t1 = np.arange(sTime, eTime, 1.0 / Fs)
y1 = A * np.cos(2 * np.pi * F1 * t1 + Phi)

# Y2 = B * cos(2 * pi * F2 * t + Phi)
B = 0.3
F2 = 15
t2 = np.arange(sTime, eTime, 1.0 / Fs)
y2 = B * np.cos(2 * np.pi * F2 * t2 + Phi)

y3 = y1 + y2

# How many samples in one cycle
nSamplesPeriod1 = int(Fs / F1)
nSamplesPeriod2 = int(Fs / F2)
nSamplesPeriod_LCM = lcm(int(nSamplesPeriod1), int(nSamplesPeriod2))

print(
    f"Number of samples per cycle N1 = {nSamplesPeriod1}, N2 = {nSamplesPeriod2}, LCM = {nSamplesPeriod_LCM}"
)

# We will plot the signals now in discrete time domain
cycles = 3

# Plot signal with dots representing samples, and connect the dots with lines
plt.figure(4)
plt.subplot(3, 1, 1)
plt.plot(t1, y1, 'o-', label='Signal 1')
plt.plot(t2, y2, 'o-', label='Signal 2')

plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal 1, 2')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(t1, y3, 'o-', label='Signal 3')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Signal 3')

plt.tight_layout()
```
