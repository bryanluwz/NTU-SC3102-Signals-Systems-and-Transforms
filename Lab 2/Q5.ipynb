5.1 


Given two discrete time systems (h1 and h2) with finite impulse response given by

h1[n] = [0.06523, 0.14936, 0.21529, 0.2402, 0.21529, 0.14936, 0.06523]

h2[n] = [-0.06523, -0.14936, -0.21529,0.7598, -0.21529, -0.14936, -0.06523]

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.arange(0, 40, 1)
num_of_samples = len(samples)

h1 = np.zeros(num_of_samples)
h2 = np.zeros(num_of_samples)
h3 = np.zeros(num_of_samples)

h1[0:7] = np.array([0.06523, 0.14936, 0.21529, 0.2402, 0.21529, 0.14936,
               0.06523])
h2[0:7] = np.array(
    [-0.06523, -0.14936, -0.21529, 0.7598, -0.21529, -0.14936, -0.06523])
h3 = np.convolve(h1, h2)[:num_of_samples]


# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 6), sharex=True)

# Plot the first signal
axs[0].plot(samples, h1, label='Signal h1', color='green', marker='o')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Signal h1')
axs[0].grid(True)

# Plot the second signal
axs[1].plot(samples, h2, label='Signal h2', color='green', marker='o')
axs[1].set_xlabel('Sample')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Signal h2')
axs[1].grid(True)

# Plot the third signal
axs[2].plot(samples, h3, label='Signal h3', color='green', marker='o')
axs[2].set_xlabel('Sample')
axs[2].set_ylabel('Amplitude')
axs[2].set_title('Signal h3')
axs[2].grid(True)

# Add a dotted line connecting the points to the x-axis for both signals
axs[0].vlines(samples, [0], h1, 'r', linewidth=0.5, linestyle='--')
axs[1].vlines(samples, [0], h2, 'r', linewidth=0.5, linestyle='--')
axs[2].vlines(samples, [0], h3, 'r', linewidth=0.5, linestyle='--')

# Add a line connecting the two signals
plt.subplots_adjust(hspace=0.5)
plt.show()
```

5.2 Input x[n] = = 𝛿[𝑛] − 2𝛿[𝑛 − 15]

```python
# Input
input_x = np.zeros(num_of_samples)
input_x[0] = 1
input_x[15] = -2

# Convolve the input with the the three signals then plot it out
y1 = np.convolve(input_x, h1)[:num_of_samples]
y2 = np.convolve(input_x, h2)[:num_of_samples]
y3 = np.convolve(input_x, h3)[:num_of_samples]

# Create subplots
fig, axs = plt.subplots(3, 1, figsize=(8, 6), sharex=True)

# Plot the first signal
axs[0].plot(samples, y1, label='Signal y1', color='green', marker='o')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Signal y1')
axs[0].grid(True)

# Plot the second signal
axs[1].plot(samples, y2, label='Signal y2', color='green', marker='o')
axs[1].set_xlabel('Sample')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Signal y2')
axs[1].grid(True)

# Plot the third signal
axs[2].plot(samples, y3, label='Signal y3', color='green', marker='o')
axs[2].set_xlabel('Sample')
axs[2].set_ylabel('Amplitude')
axs[2].set_title('Signal y3')
axs[2].grid(True)

# Add a dotted line connecting the points to the x-axis for both signals
axs[0].vlines(samples, [0], y1, 'r', linewidth=0.5, linestyle='--')
axs[1].vlines(samples, [0], y2, 'r', linewidth=0.5, linestyle='--')
axs[2].vlines(samples, [0], y3, 'r', linewidth=0.5, linestyle='--')

# Add a line connecting the two signals
plt.subplots_adjust(hspace=0.5)
plt.show()
```

5.3 

Let 𝑥1[𝑛] = 𝛿[𝑛] and 𝑥2[𝑛] = −2𝛿[𝑛 − 15]. This means that 𝑥[𝑛] = 𝑥1[𝑛] + 𝑥2[𝑛]. Determine whether the system 𝑦[𝑛] = ℎ3[𝑛] ∗ 𝑥[𝑛] is a linear system.

To show that this is a linear system we need to show that

𝑦[𝑛] = ℎ3[𝑛] ∗ 𝑥[𝑛] = ℎ3[𝑛] ∗ (𝑥1[𝑛] + 𝑥2[𝑛]) = y1[n] + y2[n]

where y1[n] = ℎ3[𝑛] ∗ 𝑥1[𝑛] and y2[n] = ℎ3[𝑛] ∗ 𝑥2[𝑛].

Plot ℎ3[𝑛] ∗ (𝑥1[𝑛] + 𝑥2[𝑛]) and y1[n] + y2[n] separately and show that they are equal, hence proving that the system is linear. 

```python
# Defie inputs x1 and x2
x1 = np.zeros(num_of_samples)
x1[0] = 1

x2 = np.zeros(num_of_samples)
x2[15] = -2

x3 = x1 + x2

# Convolve with h3
y1 = np.convolve(x1, h3)[:num_of_samples]
y2 = np.convolve(x2, h3)[:num_of_samples]
y12 = y1 + y2
y3 = np.convolve(x3, h3)[:num_of_samples]

# Plot y12 and y3
fig, axs = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

# Plot the first signal
axs[0].plot(samples, y12, label='Signal y12', color='green', marker='o')
axs[0].set_ylabel('Amplitude')
axs[0].set_title('Signal y1 + y2')
axs[0].grid(True)

# Plot the second signal
axs[1].plot(samples, y3, label='Signal y3', color='green', marker='o')
axs[1].set_xlabel('Sample')
axs[1].set_ylabel('Amplitude')
axs[1].set_title('Signal y3')
axs[1].grid(True)

# Add a dotted line connecting the points to the x-axis for both signals
axs[0].vlines(samples, [0], y12, 'r', linewidth=0.5, linestyle='--')
axs[1].vlines(samples, [0], y3, 'r', linewidth=0.5, linestyle='--')

# Add a line connecting the two signals
plt.subplots_adjust(hspace=0.5)
plt.show()
```
