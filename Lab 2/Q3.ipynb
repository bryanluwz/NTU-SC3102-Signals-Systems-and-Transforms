```python
import numpy as np
import matplotlib.pyplot as plt
```

Create input signal and the impulse response

```python
number_of_samples = 100

# Create a zero vector then fill with spikes
signal = np.zeros(number_of_samples)
signal[np.random.randint(0, number_of_samples, int(number_of_samples / 10))] = 1  # 5 spikes from 0 to number_of_samples

plt.figure(figsize=(8, 4))
plt.plot(signal)
plt.show()
```

```python
# Create boxcar kernel (10 samples wide, 6 samples boxcar width)
kernel = np.zeros(10)
kernel[2:8] = 1

plt.figure(figsize=(8, 4))
plt.plot(kernel)
plt.show()
```

Now convolve the signal with the kernel by taking the dot product of the kernel with each
time point of the signal. This can be done by creating a matrix of the kernel shifted each time
point of the signal.

```python
shifted_kernel = np.zeros((number_of_samples, number_of_samples + len(kernel) - 1))
reversed_kernel = kernel[::-1]

for i in range(number_of_samples):
    shifted_kernel[i, i:i + len(kernel)] = reversed_kernel

# We multiply the shifted kernal and the signal to perform convolution
convolved_signal = np.dot(signal, shifted_kernel)

plt.figure(figsize=(8, 4))
plt.plot(convolved_signal )
plt.ylabel("Intensity")
plt.xlabel("Time")
plt.title("Signal convolved with boxcar kernel")
plt.show()
```
